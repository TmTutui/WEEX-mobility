```python
# Use the same datatypes in columns
clean_df = raw_df.copy()
clean_df['REGETUD'] = clean_df['REGETUD'].astype(str)
clean_df['REGION'] = clean_df['REGION'].astype(str)
```

```python
# Region ou region d`etudes 84(Rhone Alpes)
clean_df = clean_df[(clean_df['REGION'] == '84') | (clean_df['REGETUD'] == '84')]
```

```python
temp = clean_df.groupby(['REGION', 'REGETUD'])[['REGETUD']].count().rename(columns={'REGETUD': 'count'})
```
